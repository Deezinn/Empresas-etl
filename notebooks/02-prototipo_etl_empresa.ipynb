### Verificando coluna por coluna

```python
# Primeira coluna -> _id

dataframe_empresa['_id'].isna().value_counts()
# False -> 67
dataframe_empresa['_id'].isnull().value_counts()
# False -> 67
dataframe_empresa['_id'].duplicated().value_counts()
# False -> 67

""" 
Espera-se que o id seja do tipo inteiro
"""

dataframe_empresa['_id'].dtype
# dtype('int64')
```

```python
# Primeira coluna -> _id

dataframe_empresa['cnpj'].isna().value_counts()
# False -> 67
dataframe_empresa['cnpj'].isnull().value_counts()
# False -> 67
dataframe_empresa['cnpj'].duplicated().value_counts()

# True -> 42
# False -> 25

# Retornou 42 cnpj duplicados, vamos entender o porquê

dataframe_empresa[dataframe_empresa['cnpj'].duplicated()]
```
