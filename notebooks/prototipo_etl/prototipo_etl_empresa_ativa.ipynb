### Verificando valores unicos para ver se tem um comportamento inesperado

```python
cnpj_com_letras = dataframe_empresa.loc[
    dataframe_empresa['cnpj'].str.contains(r'[^0-9]', na=False),
    'cnpj'
]
print(cnpj_com_letras)
```

### Infelizmente esse conjunto de dados contém cnpj com registros em hash, irei considerar isso como um ruído e separar em outro dataset

```python
dataframe_empresa[dataframe_empresa['cnpj'].isin(cnpj_com_letras)].to_csv(
    '../../data/csv/raw/EmpresasAtivasComHash.csv', index=False
)
dataframe_empresa = dataframe_empresa[~dataframe_empresa['cnpj'].isin(cnpj_com_letras)]

dataframe_empresa.head()
```

```python
# Terceira coluna -> nome_fantasia

dataframe_empresa['nome_fantasia'].isna().value_counts()

# False -> 52
# True -> 7


# Retornou algumas empresas com Nan
# Irei substituir Nan por "não informado"
texto_nao_informado = 'Não informado'

dataframe_empresa['nome_fantasia'] = dataframe_empresa['nome_fantasia'].replace('.', texto_nao_informado)
dataframe_empresa['nome_fantasia'] = dataframe_empresa['nome_fantasia'].fillna(texto_nao_informado)
dataframe_empresa[dataframe_empresa['nome_fantasia'].isna()] # Não existe mais NaN
dataframe_empresa['nome_fantasia'] = dataframe_empresa['nome_fantasia'].str.title() # Formatação para apenas a primeira letra da palavra ficar maiuscula
dataframe_empresa['nome_fantasia'].unique()
dataframe_empresa['nome_fantasia'].isnull().value_counts()

# False -> 59

dataframe_empresa['nome_fantasia'].duplicated().value_counts()

# True -> 46 
# False -> 13

dataframe_empresa[['nome_fantasia', 'cnpj', 'cnae']].duplicated().value_counts() # verificando se existe cnpj com cnae e nome duplicado

# False -> 59
dataframe_empresa['nome_fantasia'].dtype
```

```python
# Quarta coluna -> cod_logradouro

dataframe_empresa['cod_logradouro'].isna().value_counts()

# # False -> 59

dataframe_empresa['cod_logradouro'].unique()
dataframe_empresa['cod_logradouro'].isnull().value_counts()

# False -> 59

dataframe_empresa['cod_logradouro'].duplicated().value_counts()

# True -> 49 
# False -> 10
dataframe_empresa[['cod_logradouro', 'cnpj', 'cnae']].duplicated().value_counts()
# False -> 59
dataframe_empresa['cod_logradouro'].dtype
```

```python
# Quinta coluna -> nome_logradouro

dataframe_empresa['nome_logradouro'].isna().value_counts()

# False -> 59

dataframe_empresa['nome_logradouro'].unique()
dataframe_empresa['nome_logradouro'] = dataframe_empresa['nome_logradouro'].str.title()
dataframe_empresa['nome_logradouro'].unique()

dataframe_empresa['nome_logradouro'].isnull().value_counts()

# False -> 59

dataframe_empresa['nome_logradouro'].duplicated().value_counts()

# True -> 49 
# False -> 10

dataframe_empresa[['nome_logradouro', 'cnpj', 'cnae']].duplicated().value_counts()
# False -> 59

dataframe_empresa['nome_logradouro'].dtype
```

```python
# Sexta coluna -> numero_residencia

dataframe_empresa['numero_residencia'].isna().value_counts()

# True -> 35
# False -> 24

dataframe_empresa[dataframe_empresa['numero_residencia'].isna()]
dataframe_empresa['numero_residencia'] = dataframe_empresa['numero_residencia'].fillna('Não informado')
dataframe_empresa['numero_residencia'].unique()
dataframe_empresa['numero_residencia'].unique()
dataframe_empresa['numero_residencia'].isnull().value_counts()

# False -> 59

dataframe_empresa['numero_residencia'].duplicated().value_counts()

# True -> 55 
# False -> 4

dataframe_empresa[dataframe_empresa['numero_residencia'].duplicated()]

dataframe_empresa[['numero_residencia', 'cnpj', 'cnae']].duplicated().value_counts()
# False -> 59

dataframe_empresa['numero_residencia'].dtype
```
