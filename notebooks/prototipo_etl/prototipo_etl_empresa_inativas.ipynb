### Verificando valores unicos para ver se tem um comportamento inesperado

```python
cnpj_com_letras = dataframe_empresa.loc[
    dataframe_empresa['cnpj'].str.contains(r'[^0-9]', na=False),
    'cnpj'
]
print(cnpj_com_letras)
```

### Infelizmente esse conjunto de dados contém cnpj com registros em hash, irei considerar isso como um ruído e separar em outro dataset

```python
dataframe_empresa[dataframe_empresa['cnpj'].isin(cnpj_com_letras)].to_csv(
    '../../data/csv/raw/EmpresasInativasComHash.csv', index=False
)
dataframe_empresa = dataframe_empresa[~dataframe_empresa['cnpj'].isin(cnpj_com_letras)]

dataframe_empresa.head()
```
